```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["RI"]=le.fit_transform(df["RI"])
df["Na"]=le.fit_transform(df["Na"])
df["Mg"]=le.fit_transform(df["Mg"])
df["Al"]=le.fit_transform(df["Al"])
df["Si"]=le.fit_transform(df["Al"])
df["K"]=le.fit_transform(df["K"])
df["Ca"]=le.fit_transform(df["Ca"])
df["Ba"]=le.fit_transform(df["Ba"])
df["Fe"]=le.fit_transform(df["Fe"])
df.head()
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
y.value_counts()
```

```python
#oversampling;
from imblearn.over_sampling import SMOTE
over=SMOTE()
X,y=over.fit_resample(X,y)
```

```python
#scaling;
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,random_state=1,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3,weights="distance")
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="gini",max_depth=2)
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion="gini",max_depth=4,min_samples_split=4,random_state=2)
rf.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(random_state=3,n_estimators=100)
ad.fit(X_train,y_train)
```

```python
y_pred=ad.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
